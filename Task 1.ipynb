```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
for i in df.columns:
    plt.figure(figsize=(2,2))
    ax=plt.subplots(1,1)
    if df[i].dtypes=='object':
        df[i].value_counts().plot(kind='pie',autopct='%0.2f%%')
        plt.show()
    else:
        sns.histplot(df[i])
        plt.title(i)
        plt.show()
```
